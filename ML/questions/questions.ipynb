```python
import sklearn
import pandas as pd
import numpy as np
import re
import collections
import nltk
import pymorphy2
import operator
from nltk.stem.snowball import SnowballStemmer
```

```python
stop_words = []
with open("stop_words.txt", encoding="utf-8") as inp:
    m = inp.readlines()
    n = m[0].split(",")
    stop_words.append(n)
stop_words = stop_words[0]
len(stop_words)
```

```python
morph = pymorphy2.MorphAnalyzer()

def tokenize(text):
    min_length = 3
    lower = (word.lower() for word in nltk.wordpunct_tokenize(text))
    pattern = re.compile("[а-я]+")
    words = pattern.findall(" ".join(lower))
    norm_form = (morph.parse(word)[0].normal_form for word in words)
    tokens = [word for word in norm_form if word not in stop_words]
    return tokens
```

```python
def make_dict(t):
    dic = collections.Counter()
    for i in t:
        dic += collections.Counter(i)
    return dict(dic)
```

```python
text = data['quest'].apply(lambda x: tokenize(x))
text = list(text)
dic = make_dict(text)
sorted_dic = dict(sorted(dic.items(), key=operator.itemgetter(1))[::-1])
print(sorted_dic)
```
